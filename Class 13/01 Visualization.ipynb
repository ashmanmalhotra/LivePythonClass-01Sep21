```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.axvline(x=0.4, ymin=0, ymax=1, c='r', label="Warning Line", lw=5)
plt.axhline(y=0.4, xmin=0.2, xmax=0.8, c='g', label="Pass Line", lw=5)
plt.xlabel("X-Axis Label", color='red', family='Consolas', fontsize=16, weight='bold')
plt.ylabel("Y-Axis Label", color='red', family='Consolas', fontsize=16, weight='bold')
plt.title("Plot Title", color='red', family='Cooper Black', fontsize=20, weight='bold', pad=20)
plt.legend()
plt.show()
```
