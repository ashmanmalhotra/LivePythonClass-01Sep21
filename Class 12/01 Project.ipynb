```python
fr = open('height-weight.csv', 'r')
data = fr.read()
fr.close()
data
```

```python
# Convert Data to the List
data = data.split('\n')
data
```

```python
temp_list = []

for item in data:
    splitted_item = item.split('\t')
    temp_list.append(splitted_item)

print(temp_list)
```

```python
# Extract Height and Weight
height = []
weight = []
for item in temp_list[1:]:
    item_height = float(item[1])
    item_weight = float(item[2])
    height.append(item_height)
    weight.append(item_weight)
```
