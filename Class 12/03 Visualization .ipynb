# Vertical Line and Horizontal Line Plots

- axvline()
- axhline()

```python
plt.axvline(x=0.4, ymin=0, ymax=1);
```

```python
plt.axvline(x=0.4, ymin=0, ymax=1, c='green');
```

```python
plt.axvline(x=0.4, ymin=0.2, ymax=0.8, c='green');
```

```python
plt.axhline(y=0.4, xmin=0, xmax=1, c='red');
```

```python
plt.axhline(y=0.4, xmin=0.2, xmax=0.5, c='green');
```

```python
plt.axhline(y=0.2, xmin=0, xmax=1, c='red')
plt.axhline(y=0.5, xmin=0.2, xmax=0.8, c='green')
plt.axhline(y=0.7, xmin=0.2, xmax=0.5, c='blue')
```

```python
plt.axvline(x=0.2, ymin=0, ymax=1, c='red')
plt.axvline(x=0.5, ymin=0.2, ymax=0.8, c='green')
plt.axvline(x=0.7, ymin=0.2, ymax=0.5, c='blue')
```

```python
plt.axhline(y=0.2, xmin=0, xmax=1, c='red')
plt.axhline(y=0.5, xmin=0.2, xmax=0.8, c='green')
plt.axhline(y=0.7, xmin=0.2, xmax=0.5, c='blue')

plt.axvline(x=0.2, ymin=0, ymax=1, c='red')
plt.axvline(x=0.5, ymin=0.2, ymax=0.8, c='green')
plt.axvline(x=0.7, ymin=0.2, ymax=0.5, c='blue')
```

```python
cords = [0.2, 0.5, 0.7]
colors = ['red', 'green', 'blue']

for cord, color in zip(cords, colors):
    plt.axhline(y=cord, xmin=0, xmax=1, c="#631b70")
    plt.axvline(x=cord, ymin=0, ymax=1, c=color)
```

```python
cords = [0.2, 0.5, 0.7]
colors = ['red', 'green', 'blue']

for cord, color in zip(cords, colors):
    plt.axvline(x=cord, ymin=0, ymax=1, c=color)
```

```python
plt.axvline(x=0.2, ymin=0, ymax=1, c='green', label="Green Line")
plt.legend(loc=10)
```

```python
plt.axvline(x=0.2, ymin=0, ymax=1, c='green', label="Green Line")
plt.legend(loc=10)
plt.xlabel("X Axis Label")
plt.ylabel("Y Axis Label", c="red")
plt.title("Plot Title", color="red")
```

- vlines()
- hlines()

```python
plt.vlines(x=0.2, ymin=-1.8, ymax=2, color='k')
plt.hlines(y=0.2, xmin=-1.8, xmax=2, color="red")
```

```python
plt.vlines(x=0.2, ymin=-1.8, ymax=2, label="Black Line", color='k', linestyle="dashdot")
plt.hlines(y=0.2, xmin=-1.8, xmax=2, label="Red Line", color="red", linestyle="dotted")
plt.legend()
```
